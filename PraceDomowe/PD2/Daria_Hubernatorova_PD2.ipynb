```python
le1 = LabelEncoder()
le2 = LabelEncoder()
df = data[[num_feature1, num_feature2, cat_feature1, cat_feature2, label]]
df = df.dropna()
encoded_cat1 = le1.fit_transform(df[cat_feature1])
encoded_cat2 = le2.fit_transform(df[cat_feature2])
df[cat_feature1] = encoded_cat1
df[cat_feature2] = encoded_cat2
df[label] = df[label].apply(lambda x: 1 if x == 'Yes, I am Married' else 0)
df.head()
```

```python
df.describe()
```

# Model

```python
X = df.drop(label, axis=1)
y = df[label]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123)
model = XGBClassifier()
model.fit(X_train.values, y_train.values)
pred_y = model.predict(X_test.values)
score = accuracy_score(pred_y, y_test.values)
print(f'Accuracy: {score}')
```

# Wykresy ceteris paribus

```python
def ceteris_paribus(model, sample, feature, features):
    i = df.columns.get_loc(feature)
    min_value = min(features[:,i])
    max_value = max(features[:,i])
    x = np.linspace(min_value,max_value,len(np.unique(features[:,i])))
    x_to_predict = np.tile(sample,(len(np.unique(features[:,i])), 1))
    x_to_predict[:,i] = x
    y = model.predict_proba(x_to_predict)[:,1]
    sample_pred = model.predict_proba(sample)[0, 1]
    fig = plt.figure(figsize=(15, 10))
    plt.plot(x,y)
    plt.scatter(sample[0, i], sample_pred, c = 'red', s = 70)
    plt.xlabel(feature)
    plt.ylabel('Married probability')
```

```python
def pdplot(model, features, feature):
    feature_index = df.columns.get_loc(feature)
    fig = plt.figure(figsize=(15, 10))
    unique_features = np.sort(np.unique(features[:,feature_index]))
    all_proba = []
    for j in range(features.shape[0]):
        x_to_predict = np.tile(features[j],(len(unique_features),1))
        x_to_predict[:,feature_index] = unique_features
        y = model.predict_proba(x_to_predict)[:,1]
        all_proba.append(y)
        plt.plot(unique_features, y, color = '0.75', zorder = 5, linewidth = 0.5)
    plt.plot(unique_features, np.mean(all_proba, axis = 0), color = 'black',linewidth = 5 , zorder = 15)
    plt.xlabel(feature)
    plt.ylabel('Married probability')
    plt.title('Partial dependency profile')
```

```python
pdplot(model, X.values, 'w6_q21b_year')
```

```python
def mplot(model, features, feature, window):
    feature_index = df.columns.get_loc(feature)
    fig = plt.figure(figsize=(15, 10))
    unique_features = np.sort(np.unique(features[:,feature_index]))
    all_proba = []
    for j in range(features.shape[0]):
        x_to_predict = np.tile(features[j],(len(unique_features),1))
        x_to_predict[:,feature_index] = unique_features
        y = model.predict_proba(x_to_predict)[:,1]
        all_proba.append(y)
    min_value, max_value = int(min(features[:,feature_index])), int(max(features[:,feature_index]))
    m_plot = []
    for i in range(min_value, max_value):
        window_avg = np.mean(np.array(all_proba)[(features[:,feature_index] >= i - window) & (features[:,feature_index] < i + window)])
        m_plot.append(window_avg)    
    plt.plot(range(min_value,max_value), m_plot, color = 'red',linewidth = 5)
    plt.xlabel(feature)
    plt.ylabel('Married probability')
    plt.title('Marginal plot')
```

```python
mplot(model, X.values, 'w6_q21b_year', 10)
```

```python
pdplot(model, X.values, 'w6_q24_length')
```

```python
mplot(model, X.values, 'w6_q24_length', 500)
```

M-Plots avoid averaging predictions of unlikely data instances, but they mix the effect of a feature with the effects of all correlated features. 
Ceteris paribus curves for each feature are even more intuitive to understand than partial dependence plots. One line represents the predictions for one instance if we vary the feature of interest.
If the feature for which you computed the PDP is not correlated with the other features, then the PDPs perfectly represent how the feature influences the prediction on average.
For feature w6_q21b_year (year subject began romantic relationship with partner) PD plot and M plot have similar directions but PD plot shows rapid decrease in married probability. Here PD plot is more intuitive and simpler to explain, as more you know your partner, the higher is probability to be married. Correlation with other features is small, therefore PD plot highlights true feature influence even though effects of other features are not included.
For feature w6_q24_length - length of q24 how met answer in characters, which is almost uncorrelated with other features, PD plot and M plot look different. Looking at all ceteris paribus profiles, there is significant difference in each instance predictions. The difference between min and max values for this feature is high, it means that M plot with window 500 characters shows different profile beginning from 1500 characters.
