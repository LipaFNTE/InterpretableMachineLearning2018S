## Inspecting data

```python
seed = 12

df = pd.read_stata('../data/HCMST 2017 fresh sample for public sharing draft v1.1.dta')
df.head()
```

### Chosen features

```python
num_feature1 = 'w6_q16'  # how many relatives see in person each month
num_feature2 = 'time_from_met_to_rel'  # Time from meeting to relationship
cat_feature1 = 'ppgender'  # Gender
cat_feature2 = 'ppwork'  # Current Employment Status
label = 'S1'  # Are you married?
```

#### Encoding data

```python
def encode_cat(column):
    le = LabelEncoder()
    encoded_col = le.fit_transform(column)
    return le, encoded_col
```

```python
df = df[[num_feature1, num_feature2, cat_feature1, cat_feature2, label]]
df = df.dropna()

le1, encoded_cat1 = encode_cat(df[cat_feature1])
le2, encoded_cat2 = encode_cat(df[cat_feature2])
df[cat_feature1] = encoded_cat1
df[cat_feature2] = encoded_cat2
df[label] = df[label].apply(lambda x: 1 if x == 'Yes, I am Married' else 0)
df.head()
```

```python
df.describe()
```

#### Train / test split

```python
X = df.drop(label, axis=1)
y = df[label]
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=seed)
```

## Model training and validation

```python
model = XGBClassifier()
model.fit(train_X.values, train_y.values)
```

```python
pred_y = model.predict(test_X.values)
score = accuracy_score(pred_y, test_y.values)
print(f'Accuracy: {(score*100):.2f}%')
```

## Plotting

#### Plotting functions 

```python
def calculate_ICE(model, df, feature, npoints, center):
    f_min, f_max = df[feature].min(), df[feature].max()
    f_idx = df.columns.get_loc(feature)
    samples = df.values
    grid_values = np.linspace(f_min, f_max, npoints)
    all_preds = np.zeros(shape=(samples.shape[0], npoints))
    for i, val in enumerate(grid_values):
        samples[:, f_idx] = val
        preds = model.predict_proba(samples)[:, 1]
        all_preds[:, i] = preds
        
    if center:
        all_preds = all_preds - np.expand_dims(all_preds[:, 0], axis=1)
    return grid_values, all_preds


def calculate_ALE(model, df, feature, npoints):
    f_min, f_max = df[feature].min(), df[feature].max() 
    f_idx = df.columns.get_loc(feature)
    grid_values = np.linspace(f_min, f_max, 50)
    accumulated_effects = [0]
    
    for i in range(npoints-1):
        val_a, val_b = grid_values[i], grid_values[i+1]
        samples = df[(df[feature] >= val_a) & (df[feature] < val_b)].values
        samples[:, f_idx] = val_a
        if len(samples) == 0:
            accumulated_effects.append(accumulated_effects[-1])
        else:
            preds_a = model.predict_proba(samples)[:, 1]
            samples[:, f_idx] = val_b
            preds_b = model.predict_proba(samples)[:, 1]
            local_effect = np.mean(preds_b - preds_a)
            acc_loc_effect = accumulated_effects[-1] + local_effect
            accumulated_effects.append(acc_loc_effect)
    
    return grid_values, accumulated_effects


def plot_partial_dependency(model, df, features, npoints):
    if isinstance(npoints, int) and isinstance(features, str):
        npoints = [npoints]
        features = [features]
    assert len(npoints) == len(features)    
    
    fig, axes = plt.subplots(nrows=1, ncols=len(features), figsize=(15, 6))
    for i, (feature, n) in enumerate(zip(features, npoints)):
        grid_values, ice_values = calculate_ICE(model, df, feature, n, center=False)
        pd_line = np.mean(ice_values, axis=0)
        axes[i].plot(grid_values, pd_line, linewidth=4)
        axes[i].set_title(f'PD plot for {feature}')
      
    

def plot_ICE(model, df, features, npoints, center=False, show_pd=False):
    if isinstance(npoints, int) and isinstance(features, str):
        npoints = [npoints]
        features = [features]
    assert len(npoints) == len(features)    

    fig, axes = plt.subplots(nrows=1, ncols=len(features), figsize=(15, 6))
    for i, (feature, n) in enumerate(zip(features, npoints)):
        grid_values, ice_values = calculate_ICE(model, df, feature, n, center)
        for ice in ice_values:
            axes[i].plot(grid_values, ice, color='black', alpha=0.1)
        
        axes[i].set_title(f'ICE plot for {feature}')
        if show_pd:
            pd_line = np.mean(ice_values, axis=0)
            axes[i].plot(grid_values, pd_line, color='yellow', linewidth=7)
            
            
def plot_ALE(model, df, features, npoints):
    if isinstance(npoints, int) and isinstance(features, str):
        npoints = [npoints]
        features = [features]
    assert len(npoints) == len(features)   
    
    fig, axes = plt.subplots(nrows=1, ncols=len(features), figsize=(15, 6))
    for i, (feature, n) in enumerate(zip(features, npoints)):
        grid_values, ale_values = calculate_ALE(model, df, feature, n)
        axes[i].plot(grid_values, ale_values, linewidth=5)
        axes[i].set_title(f'ALE plot for {feature}')
```

### Individual Conditional Expectation Plots
#### Normal

```python
plot_ICE(model, train_X, [num_feature1, num_feature2], [50, 50], center=False, show_pd=True)
```

#### Centered

```python
plot_ICE(model, train_X, [num_feature1, num_feature2], [50, 50], center=True, show_pd=True)
```

#### Partial Dependency Plots 

```python
plot_partial_dependency(model, train_X, [num_feature1, num_feature2], [50, 50])
```

#### Accumulated Local Effect Plots

```python
plot_ALE(model, train_X, [num_feature1, num_feature2], [50, 50])
```
